```python
ventas = [
    ("Laptop", 5),
    ("Mouse", 12),
    ("Teclado", 8),
    ("Monitor", 3),
    ("USB", 20)
]

print(", ".join(str(x) for x in sorted(ventas, key = lambda x:x[1], reverse=True)[:3]))

# top3 = sorted(ventas, key = lambda x:x[1], reverse=True)[:3]
```

```python
# Ejercicio 2: Eliminar duplicados conservando orden
usuarios = [
    "ana",
    "juan",
    "ana",
    "pedro",
    "juan",
    "maria"
    ]

resultado = []
for usuario in usuarios:
    if usuario not in resultado:
        resultado.append(usuario)

print(", ".join(resultado))
print(", ".join(dict.fromkeys(usuarios)))
```

```python
# Ejercicio 3: Agrupar números pares e impares
numeros = list(range(1,11))

print("Pares:",",".join(str(n) for n in numeros if n % 2 == 0))
print("Impares:",",".join(str(n) for n in numeros if n % 2 != 0))
```
