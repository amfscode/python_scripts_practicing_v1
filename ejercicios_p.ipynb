```python
ventas = [
    ("Laptop", 5),
    ("Mouse", 12),
    ("Teclado", 8),
    ("Monitor", 3),
    ("USB", 20)
]

print(", ".join(str(x) for x in sorted(ventas, key = lambda x:x[1], reverse=True)[:3]))

# top3 = sorted(ventas, key = lambda x:x[1], reverse=True)[:3]
```

```python
# LISTA - Ejercicio 2: Eliminar duplicados conservando orden
usuarios = [
    "ana",
    "juan",
    "ana",
    "pedro",
    "juan",
    "maria"
    ]

resultado = []
for usuario in usuarios:
    if usuario not in resultado:
        resultado.append(usuario)

print(", ".join(resultado))
print(", ".join(dict.fromkeys(usuarios)))
```

```python
# LISTA Ejercicio 3: Agrupar números pares e impares
numeros = list(range(1,11))

print("Pares:",",".join(str(n) for n in numeros if n % 2 == 0))
print("Impares:",",".join(str(n) for n in numeros if n % 2 != 0))

```

```python
# 2. TUPLA - Ejercicio 4: Analizar coordenadas
puntos = [
    (2,3),
    (5,1),
    (7,8),
    (4,2),
]

mayor_x = max(puntos, key=lambda p:p[0])
mayor_y = max(puntos, key=lambda p:p[1])
print("Mayor X:", mayor_x)
print("Mayor y:", mayor_y)
```
